```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
for i in range(1,3):
    url = "https://www.amazon.in/gp/bestsellers/books/"
    r = requests.get(url)
    print(r)
```

```python
type(r.content)
```

```python
soup = BeautifulSoup(r.text,"lxml")
```

```python
soup = BeautifulSoup(r.content, "html.parser")
```

```python
container = soup.find_all('div', class_="a-column a-span12 a-text-center _cDEzb_grid-column_2hIsc")
```

```python
names = []
for item in container:
    title = item.find('div', class_="_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y")
    author = item.find('a', class_="a-size-small a-link-child")
    
    if title:
        titles = title.get_text().strip()
    else:
        titles = "N/A"
        
    if author:
        authors = author.get_text().strip()
    else:
        titles = "N/A"
        
    names.append({'Title': titles, 'Author': authors})
```

```python
print(len(names))
```

```python
rating_elements = soup.find_all('span', class_= "a-icon-alt")
```

```python
ratings = []

for rating_element in rating_elements:
        rating_text = rating_element.text.strip()
        ratings_star= rating_text.split()[0]  # Extract the numerical rating
        ratings.append(ratings_star)

print(ratings)
```

```python
rank = soup.find_all('span', class_= "zg-bdg-text")
```

```python
rank_list = []
for rank_ in rank:
    rank_num = rank_.text.strip()
    ranking = rank_num.split()[0]
    rank_list.append(ranking)
    
print(rank_list)
```

```python
bestseller_df = pd.DataFrame({
    "Rank": rank_list,
    "Title": [item['Title'] for item in names],
    "Author": [item['Author'] for item in names],
    "Rating": ratings
})

bestseller_df
```

```python
#converting to CSV file:
bestseller_df.to_csv("amazon_books.csv")
```
